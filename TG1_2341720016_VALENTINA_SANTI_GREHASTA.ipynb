3. Jelaskan kegunakan masing-masing library tersebut.

    Penjelasan:

    - PyPREP
    Digunakan untuk preprocessing sinyal EEG (Electroencephalography).
    Berguna untuk membersihkan data EEG dari noise, artifact (misalnya kedipan mata, gerakan otot), sehingga sinyal lebih siap dipakai untuk analisis.

    - Scipy
    Merupakan library ilmiah di Python yang banyak digunakan untuk komputasi numerik, statistik, optimasi, dan pemrosesan sinyal.
    Contoh: filter sinyal, FFT, integrasi, dan analisis data.

    - wandb (Weights & Biases)
    Platform untuk experiment tracking, visualisasi, dan manajemen proyek machine learning.
    Bisa digunakan untuk memantau training model, menyimpan metrik, dan membandingkan hasil eksperimen.

    - pyECG
    Library untuk memproses dan menganalisis sinyal ECG (Electrocardiography).
    Bisa digunakan untuk deteksi detak jantung, HRV (Heart Rate Variability), serta analisis kesehatan jantung.
